async tells Python:

This function can pause and resume later.
##############################################################################################

await means:

Pause this function here until the result is ready, but allow other async tasks to run.
#################################################################################################
Event Loop

The event loop is the "manager" of asyncio.

It continuously checks:

Task A ready?
Task B ready?
Task C waiting?

#################################################################################################
asyncio.run()

Starts the event loop.

asyncio.run(main())

You usually have one asyncio.run() in a program.

These are the three most important concepts in asyncio. Once you understand their relationship, asyncio becomes much easier.

Think of them like this:

Coroutine = The recipe (what should be done)
Task = A worker executing that recipe
Future = A promise that a result will be available later

A coroutine is an asynchronous function. When you call it,it does not execute immediately.

It only creates a coroutine object.

Think of it as a movie DVD.

You bought the DVD.

But the movie isn't playing yet.
