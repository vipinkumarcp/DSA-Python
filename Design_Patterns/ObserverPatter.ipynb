The Observer Pattern is a behavioral design pattern where one object (the Subject) automatically notifies multiple other objects (Observers) whenever its state changes.

It's a one-to-many relationship:

1 Subject → Many Observers
Real-world example

Imagine you subscribe to a YouTube channel.

YouTube Channel → Subject
Subscribers → Observers

When the channel uploads a new video, all subscribers receive a notification automatically.

          +----------------+
          |  YouTubeChannel|
          +----------------+
          | upload_video() |
          +----------------+
          /       |       \
         /        |        \
+----------+ +----------+ +----------+
| User A   | | User B   | | User C   |
+----------+ +----------+ +----------+

Components
1. Subject
Maintains a list of observers.
Allows observers to subscribe/unsubscribe.
Notifies all observers when something changes.
2. Observer
Defines an update() method.
Receives notifications from the subject.

```python
from abc import ABC, abstractmethod
from typing import List


# ==========================
# Subject
# ==========================
class Stock:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.observers: List[Observer] = []

    def add_observer(self, observer):
        self.observers.append(observer)

    def remove_observer(self, observer):
        self.observers.remove(observer)

    def notify_observers(self):
        for observer in self.observers:
            observer.update(self)

    def set_price(self, new_price):
        self.price = new_price
        print(f"\nStock price changed to {self.price}")
        self.notify_observers()


# ==========================
# Abstract Observer
# ==========================
class Observer(ABC):

    @abstractmethod
    def update(self, stock):
        pass


# ==========================
# Concrete Observers
# ==========================
class Dashboard(Observer):

    def update(self, stock):
        print(f"📊 Dashboard Updated: {stock.name} = ${stock.price}")


class EmailAlerts(Observer):

    def update(self, stock):
        print(f"📧 Email Alert: {stock.name} price changed to ${stock.price}")


class SmsAlerts(Observer):

    def update(self, stock):
        print(f"📱 SMS Alert: {stock.name} price changed to ${stock.price}")


# ==========================
# Usage
# ==========================
apple_stock = Stock("AAPL", 50)

dashboard = Dashboard()
email = EmailAlerts()
sms = SmsAlerts()

# Register observers
apple_stock.add_observer(dashboard)
apple_stock.add_observer(email)
apple_stock.add_observer(sms)

# Change stock price
apple_stock.set_price(60)

# Change again
apple_stock.set_price(75)

# Remove one observer
apple_stock.remove_observer(email)

print("\nAfter removing Email Alert:")

# Change again
apple_stock.set_price(90)
```

Real-Life Applications of the Observer Pattern

Stock Market Applications
    Notify dashboards, mobile apps, email, and SMS when stock prices change.
YouTube Notifications
    Notify subscribers when a creator uploads a new video.
Social Media Platforms
    Notify followers when someone posts new content.
Weather Applications
    Update mobile apps, websites, TVs, and smartwatches when weather data changes.
E-commerce Order Tracking
    Notify email service, SMS service, inventory system, and analytics when an order status changes.
Chat Applications
    Deliver new messages to all users in a chat room.
Banking Systems
    Trigger SMS alerts, email notifications, fraud detection, and transaction logging after account activity.
Smart Home Systems
    Motion detection triggers lights, alarms, security cameras, and mobile notifications.
File System Monitoring
    Notify backup services, antivirus software, loggers, and search indexers when files are created or modified.
GUI Applications
    Update buttons, labels, charts, or other UI components when data changes.
CI/CD Pipelines
    Notify Slack, email, deployment services, dashboards, and logging systems when a build completes.
Microservices & Event-Driven Architecture
    Publish events that multiple services consume independently (e.g., Kafka, RabbitMQ).
IoT (Internet of Things)
    Notify connected devices when sensor values (temperature, humidity, motion) change.
Gaming
    Update scoreboards, health bars, achievements, and multiplayer clients when game events occur.
Enterprise AI / MLOps
    Notify logging, monitoring, caching, analytics, and user interfaces when an AI model generates a response.
Cybersecurity Platforms
    Notify dashboards, SIEM tools, ticketing systems, email alerts, and audit logs when a threat is detected.
