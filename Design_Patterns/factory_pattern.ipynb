The Factory Pattern is a creational design pattern that provides a way to create objects without exposing the object creation logic to the client. 
Instead of using new directly, you ask a factory to create the object for you.

Why use the Factory Pattern?
Hides object creation logic.
Makes code easier to extend.
Reduces coupling between the client and concrete classes.
Follows the Open/Closed Principle (easy to add new types).

Real-world analogy

Imagine ordering coffee.

Without a factory: You go into the kitchen, find the ingredients, and make the coffee yourself.
With a factory: You tell the barista, "Give me a cappuccino." The barista decides how to make it.

The barista is the factory.
When should you use it?

Use a Factory Pattern when:

You need to create different types of objects based on input.
Object creation is complex.
You want to hide implementation details.
You expect to add new object types later.

Examples:

Database connection factories (MySQL, PostgreSQL, MongoDB)
Payment processors (Stripe, PayPal, Razorpay)
Notification services (Email, SMS, Push)
Cloud storage providers (AWS S3, Azure Blob, GCS)

```python
#Base pizza class (optional for shared behaviour)
from abc import ABC, abstractmethod

class Pizza(ABC):

    @abstractmethod
    def prepare(self):
        raise NotImplemented("This method should be overridden by subclass")
    
#specific pizza classes
class CheesePizza(Pizza):
    def prepare(self):
        return "preparing cheeze pizza"
    
class PerroniPizza(Pizza):
    def prepare(self):
        return "Preparing perroni pizza"

class VeggiePizza(Pizza):
    def prepare(self):
        return "Preparing Veggie Pizza"
    

#Factory class

class PizzaFactory:

    @staticmethod
    def create_pizza(pizza_type):
        if pizza_type == "cheese":
            return CheesePizza()
        elif pizza_type == "perroni":
            return PerroniPizza()
        elif pizza_type == "veggie":
            return VeggiePizza()
        else:
            raise ValueError(f"Unknown pizza type : {pizza_type}")
        

def main():
    try:
        user_input = "perroni" 
        Pizza = PizzaFactory.create_pizza(user_input)
        print(Pizza.prepare()) #output : preparing Pepproni Pizza
    except ValueError as e:
        print(e)

main()   
```
