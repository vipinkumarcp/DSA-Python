Adapter Design Pattern

The Adapter Pattern is a structural design pattern that allows two incompatible interfaces to work together.

It acts like a translator between two classes.

Definition: The Adapter Pattern converts the interface of one class into another interface that the client expects.

Real-Life Analogy

Imagine you have:

🔌 Laptop charger: USB-C
⚡ Wall socket: 3-pin

They cannot connect directly.

You use an adapter.

Wall Socket  --->  Adapter  --->  USB-C Charger


The adapter converts one interface into another

```python
from abc import ABC, abstractmethod

# Old class (already exists)
class BankService:

    def make_payment(self, amount):
        print(f"Payment of ₹{amount} processed through Bank Service")


# Target Interface (what the client expects)
class PaymentService(ABC):

    @abstractmethod
    def pay(self, amount):
        pass


# Adapter
class BankServiceAdapter(PaymentService):

    def __init__(self):
        self.bank_service = BankService()

    def pay(self, amount):
        # Translate the call
        self.bank_service.make_payment(amount)


# Client Code
payment = BankServiceAdapter()
payment.pay(1000)





```

Why Use the Adapter Pattern?
✅ Makes incompatible interfaces work together.
✅ Reuses existing or legacy code.
✅ Integrates third-party libraries without changing their code.
✅ Hides implementation differences from the client.
✅ Improves maintainability and flexibility.
✅ Follows the Open/Closed Principle.
