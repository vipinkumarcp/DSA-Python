State Design Pattern

The State Pattern is a behavioral design pattern that allows an object to change its behavior when its internal state changes.

Instead of using many if-else or switch statements, each state is represented by a separate class.

Definition: The State Pattern lets an object alter its behavior when its internal state changes, making it appear as if the object has changed its class.

Real-Life Analogy

Think about a traffic signal.

It has three states:

🔴 Red → Stop
🟡 Yellow → Wait
🟢 Green → Go

The traffic light is the same object, but its behavior changes based on its current state.

Problem Without State Pattern
class TrafficLight:

    def __init__(self):
        self.state = "RED"

    def action(self):

        if self.state == "RED":
            print("Stop")

        elif self.state == "YELLOW":
            print("Wait")

        elif self.state == "GREEN":
            print("Go")

Problems:

Large if-else blocks
Hard to maintain
Every new state requires modifying existing code

```python
from abc import ABC, abstractmethod


# -----------------------------
# State Interface
# -----------------------------
class TrafficLightState(ABC):

    @abstractmethod
    def handle(self):
        pass


# -----------------------------
# Concrete States
# -----------------------------
class RedState(TrafficLightState):

    def handle(self):
        print("🔴 RED - Stop")


class YellowState(TrafficLightState):

    def handle(self):
        print("🟡 YELLOW - Get Ready")


class GreenState(TrafficLightState):

    def handle(self):
        print("🟢 GREEN - Go")


# -----------------------------
# Context
# -----------------------------
class TrafficLight:

    def __init__(self):
        # Default state
        self.state = RedState()

    def set_state(self, state: TrafficLightState):
        self.state = state

    def show_signal(self):
        self.state.handle()


# -----------------------------
# Client Code
# -----------------------------
traffic_light = TrafficLight()

traffic_light.show_signal()

traffic_light.set_state(GreenState())
traffic_light.show_signal()

traffic_light.set_state(YellowState())
traffic_light.show_signal()

traffic_light.set_state(RedState())
traffic_light.show_signal()
```

Real-Life Examples of the State Design Pattern (Interview Points)
🚦 Traffic Light
States: Red, Yellow, Green
Behavior changes based on the current light.
🛒 E-commerce Order
States: New → Paid → Shipped → Delivered → Cancelled
Each state has different allowed actions.
🎵 Music Player
States: Playing, Paused, Stopped
Pressing the same button behaves differently depending on the current state.
🏧 ATM Machine
States: No Card, Card Inserted, PIN Verified, Transaction, Eject Card
Available operations depend on the current state.
🚪 Door Lock System
States: Open, Closed, Locked
Actions like open() or lock() depend on the door's state.
📄 Document Workflow
States: Draft, Review, Approved, Published, Archived
Editing is allowed only in certain states.
📱 Mobile Phone
States: Normal, Silent, Vibrate, Airplane Mode
Incoming call behavior changes with the mode.
🔋 Battery Management
States: Charging, Fully Charged, Low Battery, Discharging
Device behavior changes based on battery state.
🎮 Game Character
States: Idle, Running, Jumping, Attacking, Dead
Player actions depend on the current character state.
🚗 Car Engine
States: Off, Starting, Running, Stopped
Available actions change with the engine state.
🤖 AI Chatbot
States: Waiting for Input, Processing, Generating Response, Error
The chatbot behaves differently in each stage.
📞 Phone Call
States: Idle, Dialing, Ringing, Connected, On Hold, Ended
The available actions depend on the current call state.
🖨️ Printer
States: Ready, Printing, Out of Paper, Error
Printing is only possible in the "Ready" state.
🚦 Elevator
States: Idle, Moving Up, Moving Down, Door Open, Maintenance
Requests are handled differently depending on the elevator's state.
🎬 Video Player
States: Playing, Paused, Buffering, Stopped
Button actions vary with the current playback state.
Why Use the State Pattern?
✅ Eliminates large if-else or switch statements.
✅ Encapsulates each state in a separate class.
✅ Makes adding new states easy.
✅ Improves readability and maintainability.
✅ Follows the Open/Closed Principle.
