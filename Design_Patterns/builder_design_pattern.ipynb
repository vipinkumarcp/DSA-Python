The Builder Pattern is a creational design pattern that lets you construct complex objects step by step. Instead of passing many parameters to a constructor, you create an object by calling a series of methods that configure it, and then call build() to create the final object.

It's especially useful when:

An object has many optional fields.
You want to avoid constructors with long parameter lists.
You want to make object creation more readable and less error-prone.

```python
class House:

    def __init__(self,bedroom,bathroom,kitchen,garden,garage):
        self.bedroom = bedroom
        self.bathroom = bathroom
        self.kitchen = kitchen
        self.garden = garden
        self.garage = garage


    def __str__(self):
        features = [f"bedroom: {self.bedroom}",f"bathroom : {self.bathroom}",f"kitchen :{self.kitchen}",
                    f"garden :{self.garden}",f"garage:{self.garage}"       
                    ]
        return "\n".join(features)
        

class HouseBuilder:

    def __init__(self):

        self.bedroom = 1
        self.bathroom = 1
        self.kitchen = False
        self.garden = False
        self.garage = False

    def set_bedroom(self,count):
        self.bedroom = count
        return self
    
    def set_bathroom(self,count):
        self.bathroom = count
        return self
    
    def set_kitchen(self):
        self.kitchen = True
        return self

    def set_garden(self):
        self.garden = True
        return self

    def set_garage(self):
        self.garage = True
        return self

    def build(self):

        return House(self.bedroom,self.bathroom,self.kitchen,self.garage,self.garden)
    



house = (
    HouseBuilder()
    .set_bedroom(2)
    .set_bathroom(2)
    .set_kitchen()
    .set_garden()
    .set_garage()
    .build()
)

print(house)
```

Why Use the Builder Pattern?
    ✅ Creates complex objects step by step.
    ✅ Handles many optional parameters.
    ✅ Improves code readability with method chaining.
    ✅ Avoids constructors with long parameter lists.
    ✅ Makes objects easier to configure and maintain.
    ✅ Can create immutable objects.
    ✅ Validates object configuration before creation (build()).
