Strategy Design Pattern

    The Strategy Pattern is a behavioral design pattern that allows you to define a family of algorithms, encapsulate each one in a separate class, and make them interchangeable at runtime.

Instead of writing many if-else statements, you choose the appropriate strategy (algorithm) based on the situation.

Real-Life Analogy

Imagine you're using Google Maps.

You can choose different routes:

🚗 Car
🚶 Walking
🚲 Bicycle
🚌 Public Transport

The destination stays the same, but the route calculation algorithm changes.

Here:

Context: Google Maps
Strategies: Car, Walking, Bicycle, Public Transport

        Strategy (Interface)
             ▲
      ------------------
      |       |        |
PayPal   CreditCard   UPI
      |
   ShoppingCart (Context)

```python
from abc import ABC, abstractmethod

# Strategy Interface
class PaymentStrategy(ABC):

    @abstractmethod
    def pay(self, amount):
        pass


# Concrete Strategies
class CreditCardPayment(PaymentStrategy):

    def pay(self, amount):
        print(f"Payment processing ₹{amount} using Credit Card")


class PaypalPayment(PaymentStrategy):

    def pay(self, amount):
        print(f"Payment processing ₹{amount} using PayPal")


class CryptoPayment(PaymentStrategy):

    def pay(self, amount):
        print(f"Payment processing ₹{amount} using Crypto")


# Context
class PaymentProcessor:

    def __init__(self, strategy: PaymentStrategy):
        self.strategy = strategy

    def set_strategy(self, strategy: PaymentStrategy):
        self.strategy = strategy

    def process_payment(self, amount):
        self.strategy.pay(amount)


# Client
processor = PaymentProcessor(CreditCardPayment())
processor.process_payment(100)

processor.set_strategy(PaypalPayment())
processor.process_payment(250)

processor.set_strategy(CryptoPayment())
processor.process_payment(500)
        


    
```

Advantages
✅ Eliminates long if-else chains.
✅ Follows the Open/Closed Principle (open for extension, closed for modification).
✅ Easy to add new algorithms.
✅ Algorithms are reusable.
✅ Cleaner and more maintainable code.
✅ Algorithms can be switched at runtime.
Disadvantages
❌ More classes to maintain.
❌ Slightly more complex for very simple use cases.
❌ Overkill if there are only one or two algorithms.
