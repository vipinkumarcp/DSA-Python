The Singleton Pattern is a creational design pattern that ensures a class has only one instance throughout the application's lifetime and provides a global access point to that instance.

Why use the Singleton Pattern?

Use it when you need exactly one shared object, such as:

    Database connection manager
    Configuration manager
    Logger
    Cache manager
    Thread pool
    How it works
Make the constructor private so no other class can create an object.
Create a private static instance of the class.
Provide a public static method to return the single instance.

```python
#here we are implementing a flight control tower.single control tower to control alla

class ControlTower:
    _instance = None

    def __new__(cls):      
        ## check if an instance already exist  
        if cls._instance is None: 
            cls._instance = super().__new__(cls)
            print("initiializing control tower") 
        return cls._instance
    
    def manageFlight(self,Flight):
        print(f"Managing Flight {Flight}")

        
tower1 = ControlTower()
tower2 = ControlTower()

tower1.manageFlight("A123")
tower2.manageFlight("B345")

print(tower1 is tower2)
```
