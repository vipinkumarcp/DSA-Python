What is a Decorator?

A decorator is a function that adds or modifies the behavior of another function without changing its source code.

Think of it like wrapping a gift.
Original Function
        │
        ▼
Decorator
        │
        ▼
Enhanced Function

Functions are First-Class Objects

In Python, functions can:

    be assigned to variables
    be passed as arguments
    be returned from another function

def greet():
    print("Hello")
    a = greet
    a()
This feature makes decorators possible.

```python
def outer(func):

    def inner():
        
        print("before arg function call")
        func()
        print("after arg func call")

    return inner

def hello_world():
    print("hello world")


a = outer(hello_world)

a()

#Using @ syntax

@outer
def hey_world():
    print("hey world")

hey_world()
```

Decorator with Arguments

```python
def my_decor(func):

    def wrapper(*args, **kwargs):
        
        print("before func")
        func(*args, **kwargs)
        print("after func")

    return wrapper

@my_decor
def add(a,b):
    print("a+b")
   


add(1,2)


```

Practical Example 1 — Execution Time

```python
import time

def time_calc(func):
    
    def wrapper(*args,**kwargs):
        start = time.time()
        func(*args,**kwargs)
        end = time.time()

        print("executiom time", end-start)

    return wrapper

@time_calc
def hello_world():
    print("function called")
    time.sleep(2)

hello_world()
    
```
