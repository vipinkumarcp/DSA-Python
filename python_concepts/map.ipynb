map() → Transform each element.
filter() → Select elements that meet a condition.
reduce() → Combine all elements into a single value

map() is a built-in Python function that applies a function to every element of an iterable (like a list, tuple, or set) and returns a map object (an iterator).





```python
numbers = [1,2,3,4]

s = map(lambda x:x*x ,numbers)

print(list(s))

```

```python
names =["abcd","adfsfgga","ddsgee"]

s = list(filter(lambda name:len(name) > 4,names))

print(s)
```

```python
# Why is reduce() in functools?

# Because it's not considered a core built-in function in modern Python. You must import it:

num = [1,2,3,4,5]

from functools import reduce

s = reduce(lambda x,y:x+y,num)

print(s)

```
