What is an Iterator?

An iterator is an object that returns one item at a time.

It implements two methods:

__iter__()
__next__()

```python
number = [10,20,30,40]

it = iter(number )

print(next(it))
print(next(it))
```

creates an iterator.

Then

next(it)

internally calls

it.__next__()

Creating Your Own Iterator
class Counter:

    def __init__(self):
        self.num = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= 5:
            value = self.num
            self.num += 1
            return value
        raise StopIteration

What is a Generator?

A generator is a function that uses yield instead of return.

Every time next() is called, execution resumes from where it paused.

Notice:

The function doesn't restart.

It resumes after the previous yield.

```python
def counter():

    yield 1
    yield 2
    yield 3

ge = counter()

print(next(ge))
print(next(ge))
print(next(ge))

```

```python
def gen():

    for i in range(10):
        yield i

ge = gen()
print(next(ge))
print(next(ge))
print(next(ge))
```
